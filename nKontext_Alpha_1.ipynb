```python
# v1.
# # srt tests
# # turn into function
# subfile = pysubs2.load('yaeSearchTest.srt')

# v2.
# find all srts in current directory
# parse each one with pysubs2
# output ssaevents to gather filename, subtitle and start stop time
# Output json file dump
subfiles = {}
subfiles['entries'] = []
for file in os.listdir('.'):
    if file.endswith('.srt'):
        temp = pysubs2.load(file)
        for entry in temp:
            if "将軍" in entry.text: # This is where to include the search string
                subfiles['entries'].append({
                    'filename': os.path.splitext(file),
                    'start_time': entry.start/1000,
                    'end_time': entry.end/1000,
                    'text': entry.text
                })
        
with open('subfiledump.json', 'w') as outfile:
    json.dump(subfiles, outfile)
```

```python
#########################################
# using cell magic
#########################################
# %%html
# <div align="middle">
# <video width="80%" controls>
#       <source src="yaeSearchTest.mp4" type="video/mp4">
# </video></div>

#########################################
# explicitly calling the HTML renderer:
#########################################

# v1.
# from IPython.display import display, HTML, Javascript

# HTML("""

# <div align="middle">

# <script src="videosub-0.9.9.js"></script>
# <script>
# var vid = document.getElementById("myVideo");

# function getCurTime() { 
#     alert(vid.currentTime);
# } 

# function setCurTime() { 
#     vid.currentTime=633;
# } 

# </script> 

# <button onclick="getCurTime()" type="button">Get current time position</button>
# <button onclick="setCurTime()" type="button">Set time position to 5 seconds</button><br> 

# <video id="myVideo" width="80%" controls="controls">
#       <source src="yaeSearchTest.mp4" type="video/mp4">
#       <track src="yaeSearchTest.vtt" kind="subtitles" srclang="ja" label="日本語" />
# </video>



# </div>""")

# from IPython.display import display, HTML, Javascript

# with open('subfiledump.json') as json_file:
#     data = json.load(json_file)
#     for line in data['entries']:
#         if "戦争" in line['text']:    
            
#             print('Filename: ' + line['filename'][0])
#             print(f"Start: + {line['start_time']} -- End: {line['end_time']}" )
#             print('Text: ' + line['text'])
            
#             HTML("""

#             <div align="middle">

#             <script src="videosub-0.9.9.js"></script>
#             <script>
#             var vid = document.getElementById("myVideo");

#             function getCurTime() { 
#                 alert(vid.currentTime);
#             } 

#             function setCurTime() { 
#                 vid.currentTime=633;
#             } 

#             </script> 

#             <button onclick="getCurTime()" type="button">Get current time position</button>
#             <button onclick="setCurTime()" type="button">Set time position to 5 seconds</button><br> 

#             <video id="myVideo" width="80%" controls="controls">
#                   <source src="line['filename'][0]" type="video/mp4">
#                   <track src="yaeSearchTest.vtt" kind="subtitles" srclang="ja" label="日本語" />
#             </video>



#             </div>""")


#         else:
#             "Sorry, no examples found."




from IPython import display
from IPython.core.magic import register_cell_magic, Magics, magics_class, cell_magic
import jinja2

@magics_class
class JinjaMagics(Magics):
    '''Magics class containing the jinja2 magic and state'''
    
    def __init__(self, shell):
        super(JinjaMagics, self).__init__(shell)
        
        # create a jinja2 environment to use for rendering
        # this can be modified for desired effects (ie: using different variable syntax)
        self.env = jinja2.Environment(loader=jinja2.FileSystemLoader('.'))
        
        # possible output types
        self.display_functions = dict(html=display.HTML, 
                                      latex=display.Latex,
                                      json=display.JSON,
                                      pretty=display.Pretty,
                                      display=display.display)

    
    @cell_magic
    def jinja(self, line, cell):
        '''
        jinja2 cell magic function.  Contents of cell are rendered by jinja2, and 
        the line can be used to specify output type.

        ie: "%%jinja html" will return the rendered cell wrapped in an HTML object.
        '''
        f = self.display_functions.get(line.lower().strip(), display.display)
        
        tmp = self.env.from_string(cell)
        rend = tmp.render(dict((k,v) for (k,v) in self.shell.user_ns.items() 
                                        if not k.startswith('_') and k not in self.shell.user_ns_hidden))
        
        return f(rend)
        
    
ip = get_ipython()
ip.register_magics(JinjaMagics)
```
